### Feature for the displacement uniformity of the longitudinal mode

In order to evaluate the displacement uniformity of the longitudinal mode we define the metric "Norm of the Relative Displacement Deviation (NRDD)", which calculates the mean of the displacement at the output surface and the deviation at each node as a relative difference to this mean value. The deviations are squared, sumed up for all nodes. The root value gives the deviation criterion: a low value shows a great uniformity of the displacement at the output surface (a perfect longitudinal mode shape would have 0 as value). That criterion will be one of the target values for the models, since we want to predict the uniformity without calculating explicitely the displacements of the longitudinal mode.

```python
# Definition of the function to calculate the "L2 Norm of the Relative Displacement Deviation" (NRDD criteria) from the displacement vector
def norm_relative_displacement_deviation(disp):
    dev = (disp-disp.mean())/disp.mean()    
    return np.sqrt((dev**2).sum())
```

We add 4 new features in the DataFrame of design points to describe the shape of the longitudinal mode: mean, min, max and NRDD. The calculation has ot be performed for each design point, because the mode number of the longitudinal mode can be different between two design points.

```python
# Calculation of the 4 new features for the longitudinal mode
#for dp in range(1001, 1100): 
for dp in lst_dp:    
    print ('Design point:', dp)

    # Index of the current design point
    idx = df_params[df_params['dp_no'] == dp].index
    # Get number of the longitudinal mode
    id_long = df_params.loc[idx, 'mode_no_long'].item()
    col_disp = 'mode' + str(id_long)

    defs = df_defs.loc[df_defs['dp_no'] == dp, col_disp]
    
    # The modal displacements are mass normalized but the sign is undefined (can be positive or negative)
    if defs.mean() < 0:
        df_defs.loc[df_defs['dp_no'] == dp, col_disp] = -defs
        defs = -defs

    df_params.loc[idx, 'long_disp_mean'] = defs.mean()
    df_params.loc[idx, 'long_disp_max'] = defs.max()
    df_params.loc[idx, 'long_disp_min'] = defs.min()
    df_params.loc[idx, 'long_disp_nrdd'] = norm_relative_displacement_deviation(defs)
```

### Intermediate features

I called "intermediate features" the features calculated from the target variables. The goal is to understand how they correlate with the other target variables and to check later if they can be predicted. In this case, the created features are not calculated from the sonotrode geometry but from the modal properties directly. For each design point:
- the frequency ratio "freq_ratio" is created as the ratio of every frequency to the longitudinal frequency f0
- the frequency metrics "freq_fox_kap" is defined as 1/(f²-f0²), where f is the frequency of each mode (from Fox & Kapoor: Rates of change of eigenvalues and eigenvectors, 1968). It can not calculated for the longitudinal mode since f = f0
- the shape metrics "coupling_mac" calculates a correlation between the vector of one mode shape ui and an ideal longitudinal mode (uniform displacement vector uref with 1 on each component): MAC = (ui.uref)²/((ui.ui)*(uref.uref))
- the sensitivity is the ratio of "coupling_mac" and "freq_fox_kap" is calculated for each mode (except for the longitudinal mode)
- the shape metrics "coupling_long" is the value "coupling_mac" of the longitudinal mode particularly

```python
lst_modes = df_defs.columns[2:].to_list()

#for dp in range(1001, 1100):
for dp in lst_dp:
    print ('Design point:', dp)

    modes = df_modes[df_modes['dp_no'] == dp]
    disp = df_defs[df_defs['dp_no'] == dp]

    # Displacements matrix
    U = disp[lst_modes]

    item_long = df_params.loc[df_params['dp_no'] == dp, 'mode_no_long']
    if len(item_long) > 0:

        # Get number of the longitudinal mode, frequency and displacement
        id_long = item_long.item()
        freq = modes.loc[modes['mode_no'] == id_long, 'freq']

        if len(freq) > 0:
            f0 = freq.values[0]
            col_long_disp = 'mode' + str(id_long)
            u0 = disp[col_long_disp]

            # Frequency metrics
            freq_ratio = pd.DataFrame({'freq_ratio': modes['freq'].apply(lambda x: x/f0)})
            freq_fox_kap = pd.DataFrame({'freq_fox_kapoor': modes.loc[modes['mode_no'] != id_long, 'freq'].apply(lambda x: 1/(x**2-f0**2))})
        
        # Shape metrics

        # Coupling MAC
        df_dot = pd.DataFrame({'ui1': U.apply(sum, axis=0)}, index=modes.index)
        df_dot = pd.concat([df_dot, pd.DataFrame({'ui2': (U**2).apply(sum, axis=0)}, index=modes.index)], axis=1)
        ui1 = U.apply(sum, axis=0)
        ui2 = (U**2).apply(sum, axis=0)
        coupling_mac = ui1**2/(ui2*len(u0))
        df_dot = pd.DataFrame({'coupling_mac': coupling_mac}).reset_index(names='mode_name').set_index(modes.index)
        df_dp = pd.concat([freq_ratio, freq_fox_kap, df_dot], axis=1)

        # Sensitivity
        df_dp['sensitivity'] = np.abs(df_dp['coupling_mac'] * df_dp['freq_fox_kapoor'])
        df_modes.loc[df_modes['dp_no'] == dp, df_dp.columns] = df_dp
        df_params.loc[df_params['dp_no'] == dp, 'sum_sensi'] = (df_dp['sensitivity']**2).sum()
        
        # Coupling Longitudinal mode
        df_params.loc[df_params['dp_no'] == dp, 'coupling_long'] = df_dot.loc[df_dot['mode_name'] == col_long_disp, 'coupling_mac'].values[0]
```
